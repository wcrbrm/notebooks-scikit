```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
import seaborn as sns
sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")

iris.head()
```

```python
h1 = iris.hist(figsize=(8, 5))
```

```python
fig, ax = plt.subplots(figsize=(8, 1.5))  
sns.heatmap(iris.corr())
plt.xticks(rotation = 0)
nop = plt.title("Correllation")
```

```python
g = sns.pairplot(iris)
```

```python
g = sns.pairplot(iris, hue="species")
```

```python
pp0 = sns.pairplot(iris, height=3, vars=["sepal_width", "sepal_length"], hue="species", markers=["o", "s", "D"])
```

```python
pp1 = sns.pairplot(iris,
                 x_vars=["sepal_width", "sepal_length"],
                 y_vars=["petal_width", "petal_length"])
```

```python
 pp2 = sns.pairplot(iris, diag_kind="kde", hue="species")
```

```python
 pp3 = sns.pairplot(iris, kind="reg", hue="species")
```

```python
g = sns.pairplot(iris, diag_kind="kde", markers="+",
                  plot_kws=dict(s=50, edgecolor="b", linewidth=1),
                  diag_kws=dict(shade=True))
```
