```python
import numpy as np
import pandas as pd

df = pd.read_csv("./heart.csv")
df.shape
```

```python
df.head(5)
```

```python
df.columns
```

```python
# find out which columns are boolean, and do an auto-convertion
bool_cols = [col for col in df if df[col].dropna().value_counts().index.isin([0,1]).all()]
print("Boolean Features: ", bool_cols)

for col in bool_cols:
    df[col] = df[col].astype('bool')
```
