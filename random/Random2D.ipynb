```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
```

```python
# 2 sets divided by line
plt.figure(figsize=(5, 4))

x1 = 5 + np.random.standard_t(10, 100)
y1 = 5 + np.random.standard_t(10, 100)
plt.plot(x1, y1, 'go')

x2 = np.random.standard_t(10, 100)
y2 = np.random.standard_t(10, 100)
plt.plot(x2, y2, 'bo')

plt.show()
```

```python
# 2 sets divided by circle
plt.figure(figsize=(5, 4))

x1 = 4 * np.random.standard_t(10, 100)
y1 = 4 * np.random.standard_t(10, 100)
x = np.sqrt(x1*x1 + y1*y1) > 4
x1 = x1[x]
y1 = y1[x]

plt.plot(x1, y1, 'go')
x2 = np.random.standard_t(10, 50)
y2 = np.random.standard_t(10, 50)
plt.plot(x2, y2, 'bo')

plt.show()
```
