```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
```

```python
# CAUCHY DISTRIBUTION
s = np.random.standard_cauchy(1000000)
s = s[(s>-7) & (s<7)]  # truncate distribution so it plots well
plt.figure(figsize=(10,2))
plt.hist(s, bins=100, alpha=0.5, edgecolor='None')
plt.show()
```

```python
# EXPONENTIAL DISTRIBUTION
s = np.random.standard_exponential(100000)
s = s[s < 5]
plt.figure(figsize=(10,2))
plt.hist(s, bins=100, alpha=0.5, edgecolor='None')
plt.show()
```

```python
# NORMAL DISTRIBUTION
s = np.random.standard_normal(100000)
s = s[(s > -3) & (s < 3)]
plt.figure(figsize=(10,2))
plt.hist(s, bins=100, alpha=0.5, edgecolor='None')
plt.show()
```

```python
# NORMAL DISTRIBUTION
s = np.random.standard_t(10, 100000)
s = s[(s > -4) & (s < 4)]
plt.figure(figsize=(10,2))
plt.hist(s, bins=100, alpha=0.5, edgecolor='None')
plt.show()
```

```python
# GAMMA DISTRIBUTIONS
plt.figure(figsize=(10,2))

dist0 = np.random.standard_gamma(1.5, 1000000)
dist0 = dist0[dist0 < 8]
h0 = plt.hist(dist0, bins=100, color='r', alpha=0.5, edgecolor='None', density=True)

dist1 = np.random.standard_gamma(2, 1000000)
dist1 = dist1[dist1 < 8]
h1 = plt.hist(dist1, bins=100, color='g', alpha=0.5, edgecolor='None', density=True)

dist2 = np.random.standard_gamma(3, 1000000)
dist2 = dist2[dist2 < 8]
h2 = plt.hist(dist2, bins=100, color='b', alpha=0.5, edgecolor='None', density=True)

plt.show()
```
